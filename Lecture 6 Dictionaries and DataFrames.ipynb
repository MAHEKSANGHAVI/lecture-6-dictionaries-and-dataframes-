## DataFrames


```python
div_scores_dictionary = {"student_name": ["Jay", "Aaryan", "Sahil","Riya","Shubh"],
                         "math_scores": [20,56,76,78,90],
                         "chemistry_scores": [32,36,78,23,100],
                         "information_technology_scores": [34,6,88,100,45],
                         "physics_scores": [6,88,48,90,25]
                         }
                         
div_emails_dictionary = {"student_name": ["Jay", "Aaryan", "Sahil","Riya","Shubh", "Abdul"],
                           "student_email": ["Jay@gmail.com", "Aaryan@gmail.com", "Sahil@gmail.com","Riya@gmail.com","Shubh@gmail.com", "Abdual@gmail.com"]
                         }
```

```python
# setup
import pandas as pd
```

```python
#Converting the scores dictionary into a dataframe

scores_df = pd.DataFrame(div_scores_dictionary)
scores_df
```

```python
#First 3 Rows
scores_df.head(3)
```

```python
# Sample of 3 Rows (Any 3 Random Rows)
scores_df.sample(3)
```

```python
# Description of Key Statistics for All Numeric Columns
scores_df.describe()
```

```python
#Add a column called Total Score the Dataframe
scores_df["total_score"]= scores_df["math_scores"]+ scores_df["chemistry_scores"]+ scores_df["information_technology_scores"]
scores_df
```

##  Create a column called average score that gives you the average score of all subjects for each student

```python
# Add column called average score
scores_df["average_score"]= scores_df["total_score"]/4
scores_df
```

```python
insights_df = scores_df.describe().transpose()
insights_df
```
